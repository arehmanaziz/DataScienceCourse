# Hangman Game

```python
word_list = ["aardvark", "baboon", "camel"]
```

Random Word

```python
import random

index = random.randint(0, 2)
word = word_list[index]
print(word)
```

Stages

```python
stages = [
'''
 +---+
 |   |
 O   |
/|\  |
/ \  |
     |
=========
''', 

'''
 +---+
 |   |
 O   |
/|\  |
/    |
     |
=========
''', 

'''
 +---+
 |   |
 O   |
/|\  |
     |
     |
=========
''', 

'''
 +---+
 |   |
 O   |
/|   |
     |
     |
=========
''', 

'''
 +---+
 |   |
 O   |
 |   |
     |
     |
=========
''', 

'''
 +---+
 |   |
 O   |
     |
     |
     |
=========
''', 

'''
 +---+
 |   |
     |
     |
     |
     |
=========
'''
]
```

Main Game

```python
display = []
length = len(word)
for i in range(length):
    display.append("_")

print(display)


lives = 6
print("Total lives:", len(stages))

while True:
    
    guess = input("Guess a letter: ").lower()

    if guess in display:
        print(f"You have already guessed the letter '{guess}'. Pls guess again.")
        print("Word:", display)
        continue
    

    flag = False
    for i in range(length):
        if guess == word[i]:
            display[i] = guess
            flag = True

    
    if not flag:
        print("Incorrect guess:", guess)

        print(stages[lives])
        print("Remaining Lives:", lives)
        lives -= 1

        if lives == -1:
            print("You lose!")
            break
            

    print("Word:", display)


    if "_" not in display:
        print("Congratulations. You guessed all the letters.")
        break
```
